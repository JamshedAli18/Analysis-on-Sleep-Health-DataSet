```python
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '65+']
data['Age'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Calculate average sleep duration, stress level, and heart rate for each age group
age_group_analysis = data.groupby('Age')[['Sleep Duration', 'Stress Level', 'Heart Rate']].mean()
age_group_analysis
```

```python
age_group_analysis.plot()
plt.grid()
```

```python
Sleep_analysis = data.groupby('Sleep Disorder')[['Sleep Duration', 'Stress Level', 'Heart Rate']].mean()
Sleep_analysis
```

```python
Sleep_analysis.plot(kind='bar' ,color = ['Blue','Red','Green'],width = 0.1)
plt.grid(True, which='both', linestyle='--', linewidth=0.6)
plt.show()
plt.legend("best")
```
